```python
def Extract24HrTime(time_12hr):
    exc_day = time_12hr[-2:]
    exc_time = time_12hr[:-3]
    hr = int(exc_time.split(":")[0])
    mins = int(exc_time.split(":")[1])
    
    if exc_day == 'PM':
        if hr != 12:
            hr += 12
    else:
        if hr == 12:
            hr = 0
    
    return(hr, mins)

def ConvertTo12Hr(time_24):
    mins = (time_24 - int(time_24))*60
    hrs = int(time_24)
    if hrs > 12:
        day = 'PM'
        hrs -= 12
    elif hrs == 12:
        day = 'PM'
        hrs = hrs
    elif hrs == 0:
        day = 'AM'
        hrs += 12
    else:
        day = 'AM'
    return(hrs, mins, day)
```
